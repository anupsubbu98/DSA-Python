```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
```

```python
class LinkedList:
    def __init__(self,head=None):
        self.head = head
    
    def insert_at_start(self,data):
       
        node = Node(data,self.head)
        self.head = node
    
    def insert_at_end(self,data):
        
        if self.head is None:
            self.head = Node(data,None)
        
        else:
            itr = self.head
            while itr.next:
                itr = itr.next
            
            itr.next = Node(data,None)
            
    def insert_at_index(self,index,data):
        
        if index < 0 or index > self.get_length():
            raise Exception ('Not a valid index')
        
        elif index ==0 :
            self.insert_at_start(data)
            
        else:
        
            count = 0
            itr = self.head
            while itr:
                if count == index-1:
                    node = Node(data,itr.next)
                    itr.next = node
                    break
                itr = itr.next
                count+=1
                
    def remove_at_index(self,index):
        
        if index < 0 or index > self.get_length():
            raise Exception ('Not a valid index')
        
        elif index == 0:
            self.head = self.head.next
        
        else:
            count = 0
            itr = self.head
            while itr:
                if count == index-1:
                    itr.next = itr.next.next
                    break
                itr = itr.next
                count+= 1
                
    def insert_values(self,data_list):
        self.head = None
        for data in data_list :
            self.insert_at_end(data)
            
    def reverse_list(self,data_list):
        self.head = None
        for data in data_list:
            self.insert_at_start(data)
                
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr: 
            count+= 1
            itr = itr.next

        return count
    
    def display(self):
        
        if self.head is None:
            print('Linked List is empty...')
        
        else:
            itr = self.head
            llstr = ''
            while itr:
                llstr+= str(itr.data) +'-->'
                itr = itr.next
            
            print(llstr)
```

```python
if __name__ == '__main__':
    
    data_list = ['Anup','Subbu','Career','Happiness','Satisfaction','Courage']
    ll = LinkedList()
    ll.insert_at_start(40)
    ll.insert_at_start(30)
    ll.insert_at_start(20)
    ll.insert_at_start('10')
    ll.insert_at_index(4,50)
    ll.insert_at_index(5,60)
    ll.insert_at_end(70)
    ll.display()
    print("After deleting :",end=' ')
    ll.remove_at_index(5)
    ll.display()
    ll.insert_values(data_list)
    print("Given list :",end=' ')
    ll.display()
    print('After reversal :',end=' ') 
    ll.reverse_list(data_list)
    ll.display()
    print('Length of linked list is: ',ll.get_length()) 
```
